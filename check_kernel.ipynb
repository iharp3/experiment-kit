# Vanilla can do Greenland, 20 years, 0.25, hour

```python
import pandas as pd

df = pd.read_csv('/home/huan1531/experiment-kit/experiment/queries/get_raster_test_set_025H.csv')
q = df.to_records(index=False)[9]
q['start_time'] = "2004-01-01 00:00:00"
q['end_time'] = "2023-12-31 23:00:00"
q['spatial_resolution'] = 0.25
q['temporal_resolution'] = "hour"
q
```

```python
from vanilla.vanilla_get_raster_executor import VanillaGetRasterExecutor

qe = VanillaGetRasterExecutor(
    variable=q["variable"],
    start_datetime=q["start_time"],
    end_datetime=q["end_time"],
    max_lat=q["max_lat"],
    min_lat=q["min_lat"],
    min_lon=q["min_lon"],
    max_lon=q["max_lon"],
    spatial_resolution=q["spatial_resolution"],
    temporal_resolution=q["temporal_resolution"],
    aggregation=q["aggregation"],
)
qe.execute()
```
