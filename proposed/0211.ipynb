```python
from query_executor_get_raster import GetRasterExecutor

variable = "2m_temperature"
start_datetime = "1997-10-01 00:00:00"
end_datetime = "1999-05-01 00:00:00"

# Greenland
max_lat = 85
min_lat = 60
min_lon = -70
max_lon = -10

# aggregation
aggregation = "mean"


qe = GetRasterExecutor(
    variable=variable,
    start_datetime=start_datetime,
    end_datetime=end_datetime,
    min_lat=min_lat,
    max_lat=max_lat,
    min_lon=min_lon,
    max_lon=max_lon,
    temporal_resolution="hour",
    spatial_resolution=0.25,
    aggregation=aggregation,
)
res = qe.execute()
```
